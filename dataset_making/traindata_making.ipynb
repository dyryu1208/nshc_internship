* 팀장님이 보내주신 onion_share_link의 다크웹 데이터 수집
* 수동으로 약 300개의 html긁어와서 로컬 폴더에 저장
* 이전 연구데이터(txt파일)을 csv로 변환 후 데이터 추가 --> 약 5000개 추출
* test데이터셋 로직과 동일하게 중복제거 + 수동 중복제거 --> 2494개 데이터 존재

```python
import os
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup

startpath = 'C:/Users/NSHC/Desktop/unfamiliar_data'
filelist = []
for path, subdirs,files in os.walk(startpath) :        
    for filename in files:
        f = os.path.join(path,filename)               
        if os.path.isfile(f):                         
            if filename.endswith('html'):             
                filelist.append(f)                      
filelist
```

```python
len(filelist)
```

```python
html_list = []
for one_html in filelist:
    data = {"html" : "","label" : "","path" : ""}
    f = open(one_html,'r',encoding='ISO-8859-1')
    soup = BeautifulSoup(f,'html.parser')
    data['html'] = soup.text.replace('\n',' ')
    data['path'] = one_html.split('/')[-1].split('\\')[1].rstrip('.html')
    html_list.append(data)
df = pd.DataFrame(html_list)
print(df.shape)
df.head()
```

```python
df.drop_duplicates(subset=['html'],inplace=True)
df.drop_duplicates(subset=['path'],inplace=True)
df.shape
```

```python
# 팀장님이 주신 파일 추가
df_2 = pd.read_csv('file.csv')
df = pd.concat([df,df_2])
print(df.shape)
df.head()
```

```python
df.drop_duplicates(subset=['html'],inplace=True)
#df.drop_duplicates(subset=['path'],inplace=True)
df.shape
```

```python
df.to_csv('unfamiliar.csv',index=False)
```
