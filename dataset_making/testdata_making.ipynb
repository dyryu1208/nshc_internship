# 모델 채점용 데이터만들기
* WinSCP에서 onion-share-v4 폴더 내 데이터 로드
* 메인 html만을 가져오는 코드 작성

```python
# raw 폴더 내 모든 html 파일 끌어오기
import os
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
startpath = 'C:/Users/NSHC/Desktop/raw'
filelist = []
for path, subdirs,files in os.walk(startpath) :        
    for filename in files:
        f = os.path.join(path,filename)               
        if os.path.isfile(f):   
            if filename.startswith('index'):
                if filename.endswith('html'):             
                    filelist.append(f)                      
filelist
```

```python
html_list = []
for one_html in filelist:
    data = {"html" : "","label" : "","path" : ""}
    f = open(one_html,'r',encoding='ISO-8859-1')
    soup = BeautifulSoup(f,'html.parser')    
    data['html'] = soup.text.replace('\n',' ')
    data['path'] = one_html.split('/')[-1].split('\\')[2]
    html_list.append(data)
df = pd.DataFrame(html_list)
df.head()
```

# 중복제거 로직
* startswith() 적용해서 index.**.html 만 가져옴(main index)
* html / path 필드에 대해서 중복제거 수행
* 약 10100개 정도에서 1610개로 데이터 축소
* html 필드에 대해서 완벽하게 제거 되지 않았음!
* csv 파일 및 문장비교사이트 참고

```python
df.drop_duplicates(subset=['html'],inplace=True)
df.drop_duplicates(subset=['path'],inplace=True)
df.shape
```
